```python
import pandas as pd
```

```python
#load the data
file_path = "D:\\Essentials\\Code\\animation_series.csv"

df = pd.read_csv(file_path)

df.tail()
```

```python
#data cleaning

df['Cast'].fillna("Not_available", inplace = True)
df['Runtime'].fillna("0", inplace = True)
df.isna().sum()
df['No.of Votes'] = df['Number of Votes']
df.drop(columns = 'Number of Votes', inplace = True)
cols = list(df.columns)
order = cols[0:9] + [cols[-1]] + [cols[9]]
df = df[order]
df.head()
```

```python
#classification of series and movies
def function(x):
    if x.split('-')[0] == "TV":
        return "series"
    else:
        return "movie"
    

df['Type'] = df['Certificate'].astype(str).apply(function)
df.head(50)
```

```python
#Which movie or series has the longest runtime?
def runtime(x):
    return int(x.split()[0].replace(',', ''))
    
    
runtime_nos = pd.to_numeric(df['Runtime'].apply(runtime)).idxmax()
df.loc[runtime_nos, ['Title', 'Runtime']]
```

```python
#extract only the actual release year from the 'Release Year' column in the dataset
df['Release Year'] = df['Release Year'].astype(str).str.extract(r'(\d{4})', expand = False)
df['Release Year']
```

```python
#change the runtime to hrs and mins
df['hrs'] = df['Runtime'].apply(runtime) // 60
df['hrs'] = df['hrs'].astype(str) + ' hrs'
df['min'] = df['Runtime'].apply(runtime) % 60
df['min'] = df['min'].astype(str) + ' min'
df['Runtime'] = df['hrs'] + ' ' + df['min']
df.drop(columns = 'hrs', inplace = True)
df.drop(columns = 'min', inplace = True)
df['Runtime'] = df['Runtime'].str.replace('0 hrs', '')

df['Runtime']
```

```python
#What is the total gross revenue generated by all the movies in the dataset?
revenue = df['Gross Revenue'].loc[(df['Release Year'].astype(str) != '2023') & (df['Type'] == 'movie')]
revenue = revenue.str.replace(',', '')
revenue = pd.to_numeric(revenue)
print("total Gross Revenue of Movies =", revenue.sum())
```
